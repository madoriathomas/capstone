```python
from newscatcherapi import NewsCatcherApiClient

newscatcherapi = NewsCatcherApiClient(x_api_key='')

top_music = newscatcherapi.get_latest_headlines_all_pages(lang='en',
                                                    countries='us', 
                                                    topic='music',
                                                    max_page=31,
                                                    seconds_pause=1.0
                                                    )
```

```python
top_science = newscatcherapi.get_latest_headlines_all_pages(lang='en',
                                                    countries='us', 
                                                    topic='science',
                                                    max_page=100,
                                                    seconds_pause=1.0
                                                    )
```

```python
top_beauty = newscatcherapi.get_latest_headlines_all_pages(lang='en',
                                                    countries='us', 
                                                    topic='beauty',
                                                    max_page=44,
                                                    seconds_pause=1.0
                                                    )
```

```python
top_travel = newscatcherapi.get_latest_headlines_all_pages(lang='en',
                                                    countries='us', 
                                                    topic='travel',
                                                    max_page=10,
                                                    seconds_pause=1.0
                                                    )
```

```python
top_food = newscatcherapi.get_latest_headlines_all_pages(lang='en',
                                                    countries='us', 
                                                    topic='food',
                                                    max_page=35,
                                                    seconds_pause=1.0
                                                    )
```

```python
top_gaming = newscatcherapi.get_latest_headlines_all_pages(lang='en',
                                                    countries='us', 
                                                    topic='gaming',
                                                    max_page=100,
                                                    seconds_pause=1.0
                                                    )
```

```python
import pandas as pd

beauty = pd.DataFrame(top_beauty['articles'])
travel = pd.DataFrame(top_travel['articles'])
gaming = pd.DataFrame(top_gaming['articles'])
food = pd.DataFrame(top_food['articles'])
science = pd.DataFrame(top_science['articles'])
music = pd.DataFrame(top_music['articles'])

dfs = [beauty, travel, gaming, food, science, music]
new_articles = pd.concat(dfs)
```

```python
import pandas as pd

df = pd.read_csv('../data/scrape_news_articles.csv', sep=";", index_col=[0])
new_df= df[['topic', 'link', 'clean_url', 'published_date', 'title', 'language']]
new_df
```

```python
new_df= new_df.rename(columns={'clean_url': "domain", 'language':'lang'})
```

```python
new_df['topic']=new_df['topic'].replace('beauty', "BEAUTY").replace('food', "FOOD").replace('science', "SCIENCE").replace('travel', "TRAVEL").replace('gaming', "GAMING").replace('music', "MUSIC")
```

```python
new_df.to_csv('scrape_news_articles.csv', encoding='utf-8', sep=';')
```
